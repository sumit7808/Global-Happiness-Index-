```python
# Calculate descriptive statistics for Happiness Score
happiness_mean = df["Happiness Score"].mean()
happiness_median = df["Happiness Score"].median()
happiness_mode = df["Happiness Score"].mode()[0]
happiness_variance = df["Happiness Score"].var()
happiness_std = df["Happiness Score"].std()

# Calculate descriptive statistics for GDP per Capita
gdp_mean = df["Economy (GDP per Capita)"].mean()
gdp_median = df["Economy (GDP per Capita)"].median()
gdp_mode = df["Economy (GDP per Capita)"].mode()[0]
gdp_variance = df["Economy (GDP per Capita)"].var()
gdp_std = df["Economy (GDP per Capita)"].std()

# Print the results
print("Happiness Score Statistics:")
print(f"Mean: {happiness_mean:.2f}")
print(f"Median: {happiness_median:.2f}")
print(f"Mode: {happiness_mode:.2f}")
print(f"Variance: {happiness_variance:.2f}")
print(f"Standard Deviation: {happiness_std:.2f}\n")

print("GDP per Capita Statistics:")
print(f"Mean: {gdp_mean:.2f}")
print(f"Median: {gdp_median:.2f}")
print(f"Mode: {gdp_mode:.2f}")
print(f"Variance: {gdp_variance:.2f}")
print(f"Standard Deviation: {gdp_std:.2f}")

```

```python
# Group data by Region and calculate statistics
regional_stats = df.groupby("Region")[["Happiness Score", "Economy (GDP per Capita)"]].agg(
    ["mean", "median", "std"]
)

# Display the statistics
print(regional_stats)
```

```python
# Calculate the correlation matrix for relevant columns
correlation_matrix = df[["Happiness Score", "Economy (GDP per Capita)", "Family",
                         "Health (Life Expectancy)", "Freedom", "Trust (Government Corruption)", "Generosity"]].corr()

# Display the correlation matrix
print(correlation_matrix)
```
